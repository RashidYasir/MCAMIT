```python
import numpy as np

def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
# Input values (can be scalar, list, or numpy array)
x = np.array([-2, -1, 0, 1, 2])

# Apply sigmoid
output = sigmoid(x)

print("Input:", x)
print("Sigmoid Output:", output)
```

```python
#Derivative of Sigmoid
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    s = sigmoid(x)
    return s * (1 - s)

# Generate values
x = np.linspace(-10, 10, 100)

# Compute sigmoid and derivative
y = sigmoid(x)
y_deriv = sigmoid_derivative(x)

# Plot
plt.figure()
plt.plot(x, y, label="Sigmoid")
plt.plot(x, y_deriv, label="Sigmoid Derivative")
plt.xlabel("x")
plt.ylabel("Value")
plt.title("Sigmoid Function and Its Derivative")
plt.legend()
plt.grid()

plt.show()
```

What You Will Observe 📊
.  Sigmoid curve is S-shaped
.  Output range: (0, 1)
.  Derivative peaks at x = 0
.  For large positive/negative values → derivative becomes very small
.  This leads to the vanishing gradient problem
