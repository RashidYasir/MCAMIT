# 2. Activation Function
* This is the sigmoid function
* Converts values into range (0, 1)
* Used for binary classification

```python
# -----------------------------
# Activation Function (Sigmoid)
# -----------------------------
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

#Derivative used in backpropagation
#Important for gradient descent

* Key idea:

It tells how much to change weights

```python
# Derivative of sigmoid
def sigmoid_derivative(x):
    return x * (1 - x)
```

Input features:

2 inputs → 4 samples

```python
# -----------------------------
# Input Dataset (AND Gate)
# -----------------------------
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])
```

Output:

AND logic
Only [1,1] → 1

```python
y = np.array([[0], [0], [0], [1]])
```

#4. Initialize Weights & Bias
* Fixes randomness
* So results are same every time you run
#Network structure:

* Input: 2 neurons
* Hidden: 2 neurons
* Output: 1 neuron

* Weights between:
* Input → Hidden
* Shape: (2 × 2)

* Weights between:
* Hidden → Output
* Shape: (2 × 1)

* Bias for each layer
* Helps shift decision boundary

```python
# -----------------------------
# Initialize Weights & Bias
# -----------------------------
np.random.seed(1)

input_layer_neurons = 2
hidden_layer_neurons = 2
output_neurons = 1

# Weights
W1 = np.random.uniform(size=(input_layer_neurons, hidden_layer_neurons))
W2 = np.random.uniform(size=(hidden_layer_neurons, output_neurons))
# Bias
b1 = np.random.uniform(size=(1, hidden_layer_neurons))
b2 = np.random.uniform(size=(1, output_neurons))
```

#Learning rate
*  Controls how fast weights update

```python
# Learning rate
lr = 0.1
```

# 5. Training Loop
* Number of iterations

```python
# -----------------------------
# Training Loop
# -----------------------------
epochs = 100000
```

# Loop for training
# Step 1:

* Multiply input with weights
* Add bias

#Step 2:

* Apply activation

* Same process again

* Final prediction

# Difference between:

* Actual (y)
* Predicted

```python
for epoch in range(epochs):

    # -------- Forward Propagation --------

    # Hidden layer
    hidden_input = np.dot(X, W1) + b1
    hidden_output = sigmoid(hidden_input)

    # Output layer
    final_input = np.dot(hidden_output, W2) + b2
    predicted_output = sigmoid(final_input)
     # -------- Loss (Mean Squared Error) --------
    error = y - predicted_output
    # -------- Backpropagation --------

    # Output layer gradient
    d_output = error * sigmoid_derivative(predicted_output)

    # Hidden layer error
    error_hidden = d_output.dot(W2.T)
    d_hidden = error_hidden * sigmoid_derivative(hidden_output)
     # -------- Update Weights & Bias --------

    W2 += hidden_output.T.dot(d_output) * lr
    W1 += X.T.dot(d_hidden) * lr

    b2 += np.sum(d_output, axis=0, keepdims=True) * lr
    b1 += np.sum(d_hidden, axis=0, keepdims=True) * lr

```

```python
    # Print loss occasionally
    if epoch % 10000 == 0:
        loss = np.mean(np.square(error))
        print(f"Epoch {epoch}, Loss: {loss:.4f}")

```

```python
# -----------------------------
# Final Output
# -----------------------------
print("\nFinal Predictions:")
print(predicted_output)
```

#Same above example using Keras

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
```

```python
# Dataset
X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[0],[0],[1]])
```

```python
# Model
model = Sequential([
    Dense(2, input_dim=2, activation='sigmoid'),
    Dense(1, activation='sigmoid')
])
```

```python
# Compile
model.compile(optimizer='adam', loss='binary_crossentropy')
```

```python
# Train
model.fit(X, y, epochs=1000, verbose=0)
```

```python
# Predict
print("\nPredictions:")
print(model.predict(X))
```
