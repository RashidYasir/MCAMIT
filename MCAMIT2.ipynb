#1. Import Libraries

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten
from tensorflow.keras.datasets import mnist
```


* numpy в†’ used for numerical operations (arrays, matrices)
* tensorflow в†’ deep learning framework
* Sequential в†’ lets us build NN layer-by-layer (like stacking blocks)
* Dense в†’ fully connected layer (every neuron connects to all inputs)
* Flatten в†’ converts image into a 1D vector
* mnist в†’ built-in dataset of handwritten digits (0вЂ“9)

Loads dataset into:
* x_train в†’ training images
* y_train в†’ training labels
* x_test в†’ testing images
* y_test в†’ testing labels

рџ‘‰ Each image is 28Г—28 pixels

# 2. Load Dataset


```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```


* Pixel values are originally 0вЂ“255
* We scale them to 0вЂ“1

рџ‘‰ Why?

* Helps faster learning
* Prevents large values from dominating

# 3. Normalize Data

```python
x_train = x_train / 255.0
x_test = x_test / 255.0
```

#4. Convert to Binary Labels (9 vs Not 9)
* (y_train == 9) в†’ checks each label:
* If digit is 9 в†’ True
* Else в†’ False
* .astype(int) converts:
* True в†’ 1
* False в†’ 0

#рџ‘‰ Now problem becomes:

* 1 в†’ digit is 9
* 0 в†’ not 9

```python
y_train_binary = (y_train == 9).astype(int)
y_test_binary = (y_test == 9).astype(int)
```

# 5. Build Neural Network
* Flatten(input_shape=(28, 28))
* Converts image:
* 28Г—28 в†’ 784 values (1D vector)
* Neural networks need 1D input

```python
model = Sequential([
    Flatten(input_shape=(28, 28)),
    Dense(128, activation='relu'),
    Dense(1, activation='sigmoid')
])
```

* Dense(128, activation='relu')
* 128 neurons in hidden layer
* Each neuron:
* multiplies inputs with weights
* adds bias
* applies activation

рџ‘‰ ReLU:

* If value > 0 в†’ keep it
* Else в†’ 0

* Dense(1, activation='sigmoid')
* Only 1 neuron (binary classification)
* Outputs value between 0 and 1

рџ‘‰ Interpretation:

* close to 1 в†’ digit is 9
* close to 0 в†’ not 9

# 6. Compile Model

```python
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

* optimizer='adam'
* Algorithm that updates weights
* Smart version of gradient descent

* loss='binary_crossentropy'
* Measures error between:
* predicted value vs actual label
* Used for binary problems

* metrics=['accuracy']
* Tracks how many predictions are correct

# 7. Train Model

```python
model.fit(x_train, y_train_binary, epochs=5, batch_size=32)
```

* x_train в†’ input images
* y_train_binary в†’ labels (9 or not)

* epochs=5
* Model sees entire dataset 5 times

* batch_size=32
* Processes 32 images at a time

рџ‘‰ Training process:

1. Forward pass (prediction)
2. Compute loss
3. Backpropagation
4. Update weights

# 8. Evaluate Model

```python
loss, accuracy = model.evaluate(x_test, y_test_binary)
print("Test Accuracy:", accuracy)
```

#Tests model on unseen data
* Returns:
* loss в†’ error
* accuracy в†’ correctness

# 9. Make Prediction

```python
sample = x_test[0]
prediction = model.predict(sample.reshape(1, 28, 28))
```

# Take one test image
* .reshape(1, 28, 28) в†’ model expects batch input

```python
print("Prediction (probability of being 9):", prediction[0][0])
print("Actual label:", y_test[2])
```

* prediction[0][0] в†’ probability
* Compare with actual label
